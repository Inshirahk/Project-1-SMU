```python
# Import dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Add CSV
df = pd.read_csv("../Resources/mental_health_and_technology_usage_2024.csv")
df
```

```python
# Gender Count Bar Plot
df_bar = df["Gender"].value_counts()
plt.figure(figsize=(8, 6))  # Adjust figure size for clarity
plt.bar(df_bar.index, df_bar.values, color='skyblue')  # Added color for visual appeal
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution")
plt.show()
```

```python
# Gender Count Bar Plot
df_bar = df["Age"].value_counts()
plt.figure(figsize=(8, 6))  # Adjust figure size for clarity
plt.bar(df_bar.index, df_bar.values, color='skyblue')  # Added color for visual appeal
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.title("Study Age Distribution")
plt.show()
```

```python
# Group the data by Mental Health Status and calculate the average Social Media Usage Hours
average_social_media_hours = df.groupby('Mental_Health_Status')['Social_Media_Usage_Hours'].mean()
categories = ["Poor", "Fair", "Good", "Excellent"]  # Adjust order if necessary
boxplot_data = [df[df["Mental_Health_Status"] == category]["Social_Media_Usage_Hours"].dropna() for category in categories]

# Create the car chart
plt.figure(figsize=(10, 6))
average_social_media_hours.plot(kind='bar', color='lightblue')

# Add labels and title
plt.grid(True, axis='y', linestyle='--', color='gray', alpha=0.3)
plt.title("Social Media Usage Hours by Mental Health Status", fontsize=16)
plt.xlabel("Mental Health Status", fontsize=12)
plt.ylabel("Social Media Usage Hours", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group and count the data by Mental Health Status and Stress Level
grouped_data = df.groupby(['Mental_Health_Status', 'Stress_Level']).size().unstack(fill_value=0)

# Specify the desired order for Mental Health Status
mental_health_order = ["Poor", "Fair", "Good", "Excellent"]
stress_level_order = ["Low", "Medium", "High"]

# Convert the column to a categorical type with the specified order
df["Mental_Health_Status"] = pd.Categorical(df["Mental_Health_Status"], categories=mental_health_order, ordered=True)
df["Stress_Level"] = pd.Categorical(df["Stress_Level"], categories=stress_level_order, ordered=True)

# Reorder the grouped data by the specified Mental Health Status order
grouped_data = grouped_data.loc[mental_health_order, stress_level_order]

# Create a stacked bar chart
grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightblue', 'orange', 'green'])

# Add labels and title
plt.title("Mental Health Status vs Stress Level", fontsize=16)
plt.xlabel("Mental Health Status", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Stress Level")
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group and count the data by Mental Health Status and Stress Level
grouped_data = df.groupby(['Mental_Health_Status', 'Stress_Level']).size().unstack(fill_value=0)

# Specify the desired order for Mental Health Status
mental_health_order = ["Poor", "Fair", "Good", "Excellent"]
stress_level_order = ["Low", "Medium", "High"]

# Convert the column to a categorical type with the specified order
df["Mental_Health_Status"] = pd.Categorical(df["Mental_Health_Status"], categories=mental_health_order, ordered=True)
df["Stress_Level"] = pd.Categorical(df["Stress_Level"], categories=stress_level_order, ordered=True)

# Reorder the grouped data by the specified Mental Health Status order
grouped_data = grouped_data.loc[mental_health_order, stress_level_order]

# Create a stacked bar chart
grouped_data.plot(kind='bar', figsize=(10, 6), color=['lightblue', 'orange', 'green'])

# Add labels and title
plt.title("Mental Health Status vs Stress Level", fontsize=16)
plt.xlabel("Mental Health Status", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(title="Stress Level")
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group data by Mental_Health_Status and Stress_Level to calculate average Social_Media_Usage_Hours
mental_health_usage = (
    df.groupby(['Mental_Health_Status', 'Stress_Level'])['Social_Media_Usage_Hours'].mean().unstack())  # Unstack to create a table for easy plotting

# Improved Colors: Manually defined color palette
colors = ['#87CEEB', '#FFA500', '#228B22']  # Replace with preferred HEX codes

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(mental_health_usage.index))  # The label locations
width = 0.25  # Width of the bars

# Loop through each stress level to plot its bar
for i, column in enumerate(mental_health_usage.columns):
    ax.bar(
        x + i * width, 
        mental_health_usage[column], 
        width, 
        label=column,
        color=colors[i]
    )

# Add labels, title, and legend
ax.set_xlabel('Mental Health Status', fontsize=12)
ax.set_ylabel('Average Social Media Usage (Hours)', fontsize=12)
ax.set_title('Average Social Media Usage by Mental Health Status and Stress Level', fontsize=14)
ax.set_xticks(x + width)  # Adjust x-ticks to align with grouped bars
ax.set_xticklabels(mental_health_usage.index, rotation=45)
ax.legend(title='Stress Level')

plt.tight_layout()
plt.show()
```

```python
# Group data by Gender and calculate average hours on social media
gender_social_media_hours = df.groupby('Gender')['Social_Media_Usage_Hours'].mean()
print("Average Hours on Social Media by Gender:")
print(gender_social_media_hours)

# Bar chart for average social media hours by gender
plt.figure(figsize=(8, 5))
gender_social_media_hours.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Average Hours on Social Media by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Social Media Hours')
plt.show()

# Box plot for Social Media Usage by gender to show distribution
plt.figure(figsize=(10, 6))
df.boxplot(column='Social_Media_Usage_Hours', by='Gender')
plt.title('Distribution of Social Media Hours by Gender')
plt.xlabel('Gender')
plt.ylabel('Social Media Hours')
plt.show()
```

```python
# Group data by Age and calculate average social media hours
age_social_media_hours = df.groupby('Age')['Social_Media_Usage_Hours'].mean().sort_index()

# Plotting the line plot for average gaming hours by age
plt.figure(figsize=(14, 6))
plt.plot(age_social_media_hours.index, age_social_media_hours.values, marker='o', linestyle='-', color='b')
plt.title('Average Hours on Social Media by Age')
plt.xlabel('Age')
plt.ylabel('Average Hours on Social Media')
plt.grid(True, alpha=.5)
plt.xticks(age_social_media_hours.index, rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Create a new column for age groups by decade
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 19, 29, 39, 49, 59, 69], 
                         labels=['teens', '20s', '30s', '40s', '50s', '60s'])

# Group data by Age Group and Gender and calculate average gaming hours
age_group_sm_hours = df.groupby(['Age_Group', 'Gender'])['Social_Media_Usage_Hours'].mean().unstack()

# Plotting the average gaming hours by age group, split by gender
plt.figure(figsize=(12, 6))
age_group_sm_hours.plot(kind='bar', figsize=(12, 6))
plt.title('Average Hours on Social Media by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Average Hours on Social Media')
plt.grid(axis='y', alpha=.25)
plt.legend(title='Gender')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit labels properly
plt.show()
```
