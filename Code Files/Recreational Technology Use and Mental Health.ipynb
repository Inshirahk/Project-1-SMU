```python
# Import dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Add CSV
df = pd.read_csv("../Resources/mental_health_and_technology_usage_2024.csv")
df
```

```python
df.info()
```

```python
# Group the data by Mental Health Status and calculate the average Combined (Gaming and Socials) Usage Hours
average_combo_hours = df.groupby('Mental_Health_Status')['Combined_Usage_Hours'].mean()
categories = ["Poor", "Fair", "Good", "Excellent"]  # Adjust order if necessary
boxplot_data = [df[df["Mental_Health_Status"] == category]["Combined_Usage_Hours"].dropna() for category in categories]

# Create the car chart
plt.figure(figsize=(10, 6))
average_combo_hours.plot(kind='bar', color='lightblue')

# Add labels and title
plt.grid(True, axis='y', linestyle='--', color='gray', alpha=0.3)
plt.title("Recreational Technology Usage Hours by Mental Health Status", fontsize=16)
plt.xlabel("Mental Health Status", fontsize=12)
plt.ylabel("Usage Hours", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group the data by mental health status
categories = ["Poor", "Fair", "Good", "Excellent"]  # Adjust order if necessary
boxplot_data = [df[df["Mental_Health_Status"] == category]["Combined_Usage_Hours"].dropna() for category in categories]

# Create the boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(boxplot_data, labels=categories, patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'), medianprops=dict(color='blue', linewidth=2))

# Add labels and title
plt.title("Recreational Technology Usage Hours by Mental Health Status", fontsize=16)
plt.xlabel("Mental Health Status", fontsize=12)
plt.ylabel("Usage Hours", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group the data by Stress Level and calculate the average Recreational Tech Usage Hours
average_social_media_hours = df.groupby('Stress_Level')['Combined_Usage_Hours'].mean()
categories = ["Low", "Medium", "High"]  # Adjust order if necessary
boxplot_data = [df[df["Stress_Level"] == category]["Stress_Level"].dropna() for category in categories]

# Create the boxplot
plt.figure(figsize=(10, 6))
average_social_media_hours.plot(kind='bar', color='lightblue')

# Add labels and title
plt.grid(True, axis='y', linestyle='--', color='gray', alpha=0.3)
plt.title("Recreational Technology Usage Hours by Stress Level", fontsize=16)
plt.xlabel("Stress Level", fontsize=12)
plt.ylabel("Usage Hours", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group the data by mental health status
categories = ["Low", "Medium", "High"]  # Adjust order if necessary
boxplot_data = [df[df["Stress_Level"] == category]["Combined_Usage_Hours"].dropna() for category in categories]

# Create the boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(boxplot_data, labels=categories, patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'), medianprops=dict(color='blue', linewidth=2))

# Add labels and title
plt.title("Recreational Technology Usage Hours by Stress Levels", fontsize=16)
plt.xlabel("Stress Levels", fontsize=12)
plt.ylabel("Usage Hours", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Group data by Mental_Health_Status and Stress_Level to calculate average Combined_Usage_Hours
mental_health_usage = (
    df.groupby(['Mental_Health_Status', 'Stress_Level'])['Combined_Usage_Hours'].mean().unstack())  # Unstack to create a table for easy plotting

# Improved Colors: Manually defined color palette
colors = ['#87CEEB', '#FFA500', '#228B22']  # Replace with preferred HEX codes

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(mental_health_usage.index))  # The label locations
width = 0.25  # Width of the bars

# Loop through each stress level to plot its bar
for i, column in enumerate(mental_health_usage.columns):
    ax.bar(
        x + i * width, 
        mental_health_usage[column], 
        width, 
        label=column,
        color=colors[i]
    )

# Add labels, title, and legend
ax.set_xlabel('Mental Health Status', fontsize=12)
ax.set_ylabel('Average Usage (Hours)', fontsize=12)
ax.set_title('Average Recreational Technology Usage by Mental Health Status and Stress Level', fontsize=14)
ax.set_xticks(x + width)  # Adjust x-ticks to align with grouped bars
ax.set_xticklabels(mental_health_usage.index, rotation=45)
ax.legend(title='Stress Level')

plt.tight_layout()
plt.show()
```

```python
# Group data by Gender and calculate average hours on social media
gender_technology_hours = df.groupby('Gender')['Combined_Usage_Hours'].mean()
print("Average Hours on Recreational Technology by Gender:")
print(gender_technology_hours)

# Bar chart for average combined hours by gender
plt.figure(figsize=(8, 5))
gender_technology_hours.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Average Hours on Recreational Technology by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Usage Hours')
plt.show()

# Box plot for Social Media Usage by gender to show distribution
plt.figure(figsize=(10, 6))
df.boxplot(column='Combined_Usage_Hours', by='Gender')
plt.title('Distribution of Recreational Technology Hours by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Usage Hours')
plt.show()
```

```python
# Group data by Age and calculate average social media hours
age_social_media_hours = df.groupby('Age')['Combined_Usage_Hours'].mean().sort_index()

# Plotting the line plot for average gaming hours by age
plt.figure(figsize=(14, 6))
plt.plot(age_social_media_hours.index, age_social_media_hours.values, marker='o', linestyle='-', color='b')
plt.title('Average Hours on Recreational Technology by Age')
plt.xlabel('Age')
plt.ylabel('Average Hours')
plt.grid(True, alpha=.5)
plt.xticks(age_social_media_hours.index, rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Create a new column for age groups by decade
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 19, 29, 39, 49, 59], 
                         labels=['teens', '20s', '30s', '40s', '50s'])

# Group data by Age Group and Gender and calculate average gaming hours
age_group_sm_hours = df.groupby(['Age_Group', 'Gender'])['Combined_Usage_Hours'].mean().unstack()

# Plotting the average gaming hours by age group, split by gender
plt.figure(figsize=(12, 6))
age_group_sm_hours.plot(kind='bar', figsize=(12, 6))
plt.title('Average Hours on Recreational Technology by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Average Hours')
plt.grid(axis='y', alpha=.25)
plt.legend(title='Gender')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit labels properly
plt.show()
```
